```python
df = pd.read_csv('data.csv')
for x in df.index:
    if df.loc[x, "Duration"] > 120:
        df.loc[x, "Duration"] = 120
print(df.to_string())
```

```python
newdf= df.duplicated()
print(newdf.to_string())
```
